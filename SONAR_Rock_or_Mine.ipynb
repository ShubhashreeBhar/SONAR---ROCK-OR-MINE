Separating data and labels

```python
X=data.drop(columns=60,axis=1)
Y=data[60]
```

Train and Test Data

```python
X_train,X_test,Y_train,Y_test=train=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

MODEL EVALUATION

```python
X_train_prediction=model.predict(X_train)
training_accuracy=accuracy_score(X_train_prediction,Y_train)
print(training_accuracy)

X_test_prediction=model.predict(X_test)
training_test_accuracy=accuracy_score(X_test_prediction,Y_test)
print(training_test_accuracy)
```

MODELING

```python
classifiers = [
    ('Logistic Regression', LogisticRegression(random_state=42)),
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),
    ('K-Nearest Neighbors', KNeighborsClassifier())
]

for clf_name, clf in classifiers:
    clf.fit(X_train, Y_train)
    y_pred = clf.predict(X_test)
    f1 = f1_score(Y_test, y_pred, average='weighted')  # Using weighted average for multi-class classification
    print(f'{clf_name}: F1 Score = {f1:.2f}')
    print(f'{clf_name} Classification Report:\n{classification_report(Y_test, y_pred)}')
    print('---------------------------------------------------')
```
